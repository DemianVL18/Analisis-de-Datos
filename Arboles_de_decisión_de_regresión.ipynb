```python
# Importar bibliotecas necesarias para visualización, modelado y métricas
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split  # Para dividir los datos
from sklearn.tree import DecisionTreeRegressor         # Modelo de árbol de regresión
from sklearn.metrics import mean_squared_error, r2_score  # Métricas de evaluación
import pandas as pd
import numpy as np
import math

# Importar función para generar datos sintéticos de regresión
from sklearn.datasets import make_regression

# Generar un conjunto de datos con 1000 muestras, 3 características y algo de ruido
# X contiene las características (features), y contiene la variable objetivo (target)
X, y = make_regression(n_samples=1000, n_features=3, noise=0.1, random_state=42)
```

```python
print (X)
```

```python
# Dividir los datos en entrenamiento (70%) y prueba (30%) de forma aleatoria
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Crear un modelo de árbol de regresión con una profundidad máxima de 3
# Este modelo aprenderá a predecir y a partir de los datos de entrenamiento
tree_regressor = DecisionTreeRegressor(max_depth=3)
tree_regressor.fit(X_train, y_train)  # Entrenamiento del modelo
```

```python
# Realizar predicciones sobre el conjunto de prueba usando el modelo entrenado
y_pred = tree_regressor.predict(X_test)
print (y_pred)
```

```python
# Evaluar el rendimiento del modelo con el error cuadrático medio (MSE)
# Un valor más bajo indica mejor desempeño
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", mse)

# Calcular la raíz del MSE para interpretar el error en la misma escala que y
print("Raíz del error cuadrático medio:", math.sqrt(mse))

# Mostrar estadísticas básicas del conjunto de prueba para contextualizar los errores
print("Valor medio de y en el conjunto de prueba:", y_test.mean())
print("Valor mínimo:", y_test.min())
print("Valor máximo:", y_test.max())

# Calcular el coeficiente de determinación R2
# Indica qué tan bien el modelo explica la variabilidad de los datos (1.0 es perfecto)
r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R2):", r2)
```

```python
# Generar un conjunto de datos con 1000 muestras, 3 características y algo de ruido
# X contiene las características (features), y contiene la variable objetivo (target)
X, y = make_regression(n_samples=1000, n_features=3, noise=0.1, random_state=42)
# Dividir los datos en entrenamiento (70%) y prueba (30%) de forma aleatoria
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Crear un modelo de árbol de regresión con una profundidad máxima de 5
# Este modelo aprenderá a predecir y a partir de los datos de entrenamiento
tree_regressor = DecisionTreeRegressor(max_depth=5)
tree_regressor.fit(X_train, y_train)  # Entrenamiento del modelo
# Realizar predicciones sobre el conjunto de prueba usando el modelo entrenado
y_pred = tree_regressor.predict(X_test)
# Evaluar el rendimiento del modelo con el error cuadrático medio (MSE)
# Un valor más bajo indica mejor desempeño
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", mse)

# Calcular la raíz del MSE para interpretar el error en la misma escala que y
print("Raíz del error cuadrático medio:", math.sqrt(mse))

# Mostrar estadísticas básicas del conjunto de prueba para contextualizar los errores
print("Valor medio de y en el conjunto de prueba:", y_test.mean())
print("Valor mínimo:", y_test.min())
print("Valor máximo:", y_test.max())

# Calcular el coeficiente de determinación R2
# Indica qué tan bien el modelo explica la variabilidad de los datos (1.0 es perfecto)
r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R2):", r2)
```
