```python
# Importar libreras necesarias
from sklearn.datasets import make_blobs  # Para generar datos sintticos con estructura de clusters
from sklearn.cluster import KMeans       # Algoritmo de clustering K-Medias
import matplotlib.pyplot as plt          # Para visualizacin grfica

# Para este ejercicio, generamos un conjunto de datos sintticos que simula una situacin tpica de clustering,
#donde los puntos estn agrupados en torno a centros definidos. Usamos make_blobs porque permite controlar el nmero de muestras,
#centros (grupos), dispersin y aleatoriedad, facilitando as la validacin visual del algoritmo K-Medias.

# Generar un conjunto de datos sintticos para clustering con 300 muestras, 2 caractersticas y 4 centros (grupos)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=2.0, random_state=0)
# Resultado: X contiene las caractersticas (2 columnas), organizadas en torno a 4 grupos generados aleatoriamente

# Una vez generados los datos, aplicamos el algoritmo K-Medias, que busca dividir los puntos en 4 grupos (como se defini en los datos),
#minimizando la distancia entre los puntos y sus respectivos centroides. El ajuste del modelo (fit) permite que el algoritmo calcule
#las posiciones ptimas de estos centroides y asigne cada punto a su grupo ms cercano.

# Crear el modelo de K-Medias para agrupar los datos en 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0)
# Resultado: Se configura el modelo de clustering K-Medias para encontrar 4 grupos distintos

# Entrenar el modelo con los datos sintticos
kmeans.fit(X)
# Resultado: El modelo calcula los centroides ptimos y asigna a cada muestra una etiqueta de grupo

# Obtener las etiquetas de grupo para cada muestra
labels = kmeans.labels_
# Resultado: Cada valor indica a qu cluster pertenece la muestra correspondiente

#Los centroides son los puntos centrales de cada grupo. Representan el promedio de todas las muestras asignadas a ese cluster.
#En el grfico, se muestran con una 'X' roja, indicando el punto central de cada grupo identificado por el algoritmo.

# Obtener las coordenadas de los centroides de cada grupo
centers = kmeans.cluster_centers_
# Resultado: Arreglo con las posiciones (centros) de cada uno de los 4 clusters encontrados

# Visualizar los puntos agrupados y los centroides
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')  # Muestra los puntos coloreados por grupo
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroides')  # Muestra los centroides
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.title('Clustering con K-Medias (4 grupos)')
plt.legend()
plt.show()
# Resultado: Se genera un grfico donde se visualizan los clusters formados y los centroides en rojo

# Filtrar las muestras del grupo 2
filtro = labels == 2
X[filtro]
# Resultado: Se obtiene un subconjunto de datos correspondiente al grupo nmero 2

# Mostrar las coordenadas de los 4 centroides
for i, centroide in enumerate(centers):
    print(f"Centroide {i}: {centroide}")
# Resultado: Se visualizan las coordenadas del primer centroide (grupo 0)
```
